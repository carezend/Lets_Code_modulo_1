```python
import time

file = open(r'C:\\Users\\rezen\\Downloads\\contatos.csv', 'r')
novo = open(r'C:\\Users\\rezen\\Downloads\\contatos_ordenados.csv', 'w')


#Funções

#Leitura do arquivo csv.

lista2 = []

def ordenar ():
    
    global lista2
    
    for linha in file:
        lista = linha.strip().split(';') 

#ordenar as informações na ordem correta [nome, telefone, email]

        if '@' in lista[0]: 
            lista.insert(3, lista[0]) 
            lista.remove (lista[0]) 

        if '@' in lista[1]:
            lista.insert(3, lista[1])
            lista.remove (lista[1])

        if lista[0].isdigit() == True: 
            lista.insert(2, lista[0])
            lista.remove (lista[0])

        if lista[2].isdigit() == True:
            lista.insert(2, lista[2])
            lista.remove (lista[2])

        if lista[1].isalpha() == True: 
            lista.insert(1, lista[1])
            lista.remove(lista[1])

        if lista[2].isalpha() == True:
            lista.insert(1, lista[2])
            lista.remove(lista[2])

        lista2.append(lista)

    lista2 = sorted(lista2)

    file.close()

def cabecalho():
    global lista2
    novo.write (' PyCoders Ltda ')
    novo.write("\n")
    novo.write ('LISTA DE CONTATOS DOS FORNECEDORES')
    novo.write("\n")
    novo.write ('-----------------------------------------------------')
    novo.write("\n")
    novo.write ('NOME;TELEFONE;E-MAIL')
    novo.write("\n")

def arquivo_csv ():
    global lista2
    for linha in lista2:
        linha = str(linha).strip('[]').replace("'", "")
        linha = linha.replace (',', ';')
        novo.write(linha)
        novo.write('\n')
    
    novo.close()
    
def adiciona():
    novo = open(r'C:\\Users\\rezen\\Downloads\\contatos_ordenados.csv', 'a')
    
    contato = input('Digite um contato, seu telefone (com DD), seu e-mail, separado por (,) ')
    
    contato1 = contato.split(', ')
    
    if 10 <= len(contato1[1]) <= 11 and contato1[1].isdigit() == True:
        
        if contato1[2].endswith('.com') == True or contato1[2].endswith('.com.br') == True and contato1[2].find('@') >= 1:

            lista2.append(contato1)
            contato1 = str(contato1)
            novo.write(contato1.strip('[]').replace (',', ';').replace("'", ""))
            novo.write('\n')
            print('Adicionando...')
            time.sleep(2)
            print('Adicionado com sucesso!')
            novo.close()
        else:
            print('E-mail inválido!')
    else:
        print('Telefone inválido!')

def busca():
    global lista2
    informacao = input('Digite um nome, telefone ou e-mail para buscar: ')
    print('Buscando...')
    time.sleep(2)
    for contato in lista2:
        if informacao in contato:
            print('Nome: {}, Telefone: {}, E-mail: {}'.format(contato[0], contato[1], contato[2]))
            break
    else:
        print('Não encontrado')
    
        
            
def remove():
    global lista2
    informacao = input('Digite um contato, telefone ou e-mail para remover: ')
            
    for contato in lista2: 
        if informacao in contato:
            indice = lista2.index(contato)
            print('Removendo...')
            time.sleep(2)        
            print('Removido com sucesso')
            del lista2[indice]
            break
    else:
        print('Não encontrado')
        

    for linha in lista2:
        linha = str(linha).strip('[]').replace("'", "")
        linha = linha.replace (',', ';')
        novo.write(linha)
        novo.write('\n')
            
    novo.close()

#Criar a Classe Contato (atributos nome, telefone, email)

class Contato:
    def __init__(self, nome, telefone, email):
        self.nome = nome
        self.telefone = telefone
        self.email = email

lista_contato = []
for i in lista2:
    x = Contato(i[0], i[1], i[2])
    lista_contato.append(x)
    
#print(lista_contato[0].nome)


try:

    cmd = int(input('''Digite um dos seguintes comandos: 
                                1 - para adicionar contato
                                2 - para remover contato
                                3 - para buscar contato: '''))

    if cmd == 1:
        ordenar()
        cabecalho()
        arquivo_csv()
        adiciona()

    elif cmd == 3:
        ordenar()
        busca()

    elif cmd == 2:
        ordenar()
        cabecalho()
        remove()

    elif (cmd == 0 or cmd > 3):
        print('Operação inválida')

except ValueError:
    print('Operação inválida')
```
